```python
#Import libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
from matplotlib import figure

plt.style.use('ggplot')
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
#Read the file
df = pd.read_csv(r'D:\NewYorkAirBnB.csv')
```

```python
df.sort_values(by=['price'], inplace=False, ascending=False)
```

```python
#Looking at the correlation
df.corr() #pearson, kendall, spearman
```

```python
correlation_matrix=df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
```

```python
#List top ten host_id with the most values counts and Room Type Counts
host=df['host_id'].value_counts()[:10]
host
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x= host.index, y =host.values, order=host.index)
plt.title('Top 10 popular host in NewYork AirBnB')
plt.xlabel('host id')
plt.ylabel('Feature Counts')
```

```python
room_type=df["room_type"].value_counts()
```

```python
labels=room_type.index
values=room_type.values
plt.figure(figsize=(8,6))
plt.pie(values,labels=labels,autopct='%1.2f%%')
plt.title("Type of Rooms.")
```

```python
#What is the mean price and avg price of each neighbourhood_group of a room?
neighbor_group=[i for i in df['neighbourhood_group'].unique()]
neighbor_group
```

```python
mean_price=df['price'].mean()
mean_price
```

```python
neighbor_mean = []
for i in neighbor_group:
    neighbor_mean.append(df[df['neighbourhood_group'] == i]['price'].mean())

neighbor_mean
```

```python
neighbor_avgprice = pd.DataFrame({'neighbour_group':neighbor_group,
             'avg_price': neighbor_mean,
                           'mean_price': mean_price})
neighbor_avgprice
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(x = 'neighbour_group', height = 'avg_price', data = neighbor_avgprice)
ax2.plot('neighbour_group', 'mean_price', data = neighbor_avgprice, color = 'black')
```

```python
#Plot NYC map based on neighbourhood_group, room_type
plt.figure(figsize = (12,5))
sns.scatterplot(x = df['long'], y = df['lat'], hue = df['neighbourhood_group'])
plt.title('NewYork AirBNB Map based on Neighbourhood_group ')
```

```python
plt.figure(figsize = (12,5))
sns.scatterplot(x = df['long'], y = df['lat'], hue = df['room_type'])
plt.title('NewYork AirBNB Map based on Room_type ')
```
